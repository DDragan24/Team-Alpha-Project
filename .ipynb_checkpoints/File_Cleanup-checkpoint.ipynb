```python
# Import dependencies
import pandas as pd
from pymongo import MongoClient
import json
import requests 
from pathlib import Path
from pprint import pprint
```

```python
# Store filepath in a variable
CMPD_File = Path("Resources/CMPD_Homicide.csv")
```

```python
# reads the csv file and converts into Dataframe
CMPD_File_df = pd.read_csv(CMPD_File, encoding="ISO-8859-1")
```

```python
CMPD_File_df.head()
```

```python
# names of all columns
for col in CMPD_File_df:
    print(col)
```

```python
# deleting all unwanted columns
CMPD_File_df = CMPD_File_df[["CRIME_REPORT_ID", 
                            "AGE", 
                            "GENDER",
                            "RACE_ETHNICITY",
                            "CLEARANCE_STATUS",
                            "WEAPON", "LOCATION",
                            "CMPD_PATROL_DIVISION",
                            "CIRCUMSTANCES",
                            "CALENDAR_YEAR",
                            "CALENDAR_MONTH"]]
CMPD_File_df.head()
```

```python
# essentially combines calendar_year and month to get new column Date 
CMPD_File_df['DATE'] = CMPD_File_df['CALENDAR_MONTH'].astype(str) +"-" + CMPD_File_df['CALENDAR_YEAR'].astype(str)
CMPD_File_df.head()
```

```python
# removes year and month
CMPD_File_df = CMPD_File_df[["CRIME_REPORT_ID", 
                            "AGE", 
                            "GENDER",
                            "RACE_ETHNICITY",
                            "CLEARANCE_STATUS",
                            "WEAPON", "LOCATION",
                            "CMPD_PATROL_DIVISION",
                            "CIRCUMSTANCES",
                            "DATE"]]
CMPD_File_df.head()
```
