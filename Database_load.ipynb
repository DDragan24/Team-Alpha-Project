```python
# import dependacies
from pymongo import MongoClient
```

```python
# create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# make reference to a new database names CMPD with collection names HOMOCIDES
db = mongo.CMPD
homocides = db.HOMOCIDES
```

```python
# query the homocides
results = homocides.find()
for crime in results:
    print(crime)
```
